```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np

# Load the .mat file using scipy.io.loadmat
mat_data = scipy.io.loadmat('H_trix.mat')

# Check available keys in the file
print("Available keys in the file:", [key for key in mat_data.keys() if not key.startswith('__')])
```

```python

# Load the NTN matrix
matrix = mat_data['1_H_true_target']

print(f"Matrix shape: {matrix.shape}")
print(f"Matrix dtype: {matrix.dtype}")
print(f"Is complex: {np.iscomplexobj(matrix)}")
```

```python

matrix_slice = matrix[0, :, :, 0]

# Plot the real part of the selected slice
plt.figure(figsize=(10, 6))
plt.imshow(matrix_slice, aspect='auto', cmap='viridis')
plt.colorbar(label='Real Value')
plt.title('Real Part of NTN Matrix [n, :, :] - Shape: (132, 14)')
plt.xlabel('Columns (14)')
plt.ylabel('Rows (132)')
plt.show()

print(f"Selected slice shape: {matrix_slice.shape}")
print(f"Data type: {matrix_slice.dtype}")
```
